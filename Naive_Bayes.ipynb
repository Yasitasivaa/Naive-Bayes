## Fungsi Naive Bayes

```python
def naive_bayes_prediction(AreaManager, Kota, JenisKelamin, RataRataUmurPenyewa, RataRataBiayaSewa):
    persentase_AreaManager = {'adp': {'Paid': persen_adp_Paid, 'Out Standing': persen_adp_OutStanding},
                              'ags': {'Paid': persen_ags_Paid, 'Out Standing': persen_ags_OutStanding},
                              'bcp': {'Paid': persen_bcp_Paid, 'Out Standing': persen_bcp_OutStanding}}

    persentase_Kota = {'Bandung': {'Paid': persen_Bandung_Paid, 'Out Standing': persen_Bandung_OutStanding},
                       'Cimahi': {'Paid': persen_Cimahi_Paid, 'Out Standing': persen_Cimahi_OutStanding}}

    persentase_JenisKelamin = {'Pria': {'Paid': persen_Pria_Paid, 'Out Standing': persen_Pria_OutStanding},
                               'Wanita': {'Paid': persen_Wanita_Paid, 'Out Standing': persen_Wanita_OutStanding}}

    persentase_RataRataUmurPenyewa = {'<40': {'Paid': persen_Kurangdari40_Paid, 'Out Standing': persen_Kurangdari40_OutStanding},
                                      '>40': {'Paid': persen_Lebihdari40_Paid, 'Out Standing': persen_Lebihdari40_OutStanding}}

    persentase_RataRataBiayaSewa = {'Sewa < 500000': {'Paid': persen_Kurangdari50_Paid, 'Out Standing': persen_Kurangdari50_Outstanding},
                                    'Sewa = 500000': {'Paid': persen_Samadengan50_Paid, 'Out Standing': persen_Samadengan50_Outstanding},
                                    'Sewa > 500000': {'Paid': persen_Lebihdari50_Paid, 'Out Standing': persen_Lebihdari50_Outstanding}}

    # Menghitung prediksi untuk kelas 'Paid'
    prediksi_Paid = (
            persen_Paid *
            persentase_AreaManager.get(AreaManager, 0).get('Paid', 0) *
            persentase_Kota.get(Kota, 0).get('Paid', 0) *
            persentase_JenisKelamin.get(JenisKelamin, 0).get('Paid', 0) *
            persentase_RataRataUmurPenyewa.get(RataRataUmurPenyewa, 0).get('Paid', 0) *
            persentase_RataRataBiayaSewa.get(RataRataBiayaSewa, 0).get('Paid', 0)
    )

    # Menghitung prediksi untuk kelas 'Out Standing'
    prediksi_OutStanding = (
            persen_OutStanding *
            persentase_AreaManager.get(AreaManager, 0).get('Out Standing', 0) *
            persentase_Kota.get(Kota, 0).get('Out Standing', 0) *
            persentase_JenisKelamin.get(JenisKelamin, 0).get('Out Standing', 0) *
            persentase_RataRataUmurPenyewa.get(RataRataUmurPenyewa, 0).get('Out Standing', 0) *
            persentase_RataRataBiayaSewa.get(RataRataBiayaSewa, 0).get('Out Standing', 0)
    )

    print(f'Area Manager: {AreaManager}')
    print(f'Kota: {Kota}')
    print(f'Jenis Kelamin: {JenisKelamin}')
    print(f'Rata-Rata Umur Penyewa: {RataRataUmurPenyewa}')
    print(f'Rata-Rata Biaya Sewa: {RataRataBiayaSewa}')
    print(f'Persentase Paid: {prediksi_Paid}')
    print(f'Persentase Out Standing: {prediksi_OutStanding}')

    prediksi = max(prediksi_Paid, prediksi_OutStanding)
    if prediksi == prediksi_Paid:
        return 'Paid'
    else:
        return 'Out Standing'


AreaManager = input('Masukkan Area Manager (adp/ags/bcp): ')
Kota = input('Dari mana Kota anda? (Bandung/Cimahi): ')
JenisKelamin = input('Apa Jenis Kelamin Anda? (Pria/Wanita): ')
RataRataUmurPenyewa = input('Berapa Umur Anda? (<40/>40): ')
RataRataBiayaSewa = input('Berapa Rata-Rata Biaya Sewa Anda? (Sewa < 500000/Sewa = 500000/Sewa > 500000): ')

kelas_prediksi = naive_bayes_prediction(AreaManager, Kota, JenisKelamin, RataRataUmurPenyewa, RataRataBiayaSewa)
print("Prediksi Status Pembayaran: ", kelas_prediksi)
```
